```python
f = open('.\sr28\FOOD_DES.txt', 'r')
lines = f.readlines()
f.close()
print lines[0:5]
```

```python
tmp = []
for l in lines:
    tmp.append(l.split('~'))
print tmp[0]
```

```python
#collect all data from several files into one DataFrame
df = pd.DataFrame({'NDB_No' : map(int, zip(*tmp)[1]), 
                    'FdGrp_No' : map(int, zip(*tmp)[3]),
                    'FullName' : zip(*tmp)[5],
                    'Name' : zip(*tmp)[7]})
df = df.set_index('NDB_No')
df[0:10]
```

```python
#get nutrition data codes
nut_code = {}
f = open('.\sr28\NUTR_DEF.txt', 'r')
lines = f.readlines()
f.close()
for l in lines:
    tmp = l.split('~')
    nut_code[tmp[5]] = tmp[1]
print 'Water: ', nut_code['WATER'], \
    'Protein: ', nut_code['PROCNT'], \
    'Total lipid (fat): ', nut_code['FAT'], \
    'Energy: ', nut_code['ENERC_KCAL'], \
    'Carbohydrate:', nut_code['CHOCDF'], \
    'Fiber: ', nut_code['FIBTG'], \
    'Sugar: ', nut_code['SUGAR']
```

```python
useful_codes = {nut_code['WATER']      : 'Water', 
                nut_code['PROCNT']     : 'Protein',
                nut_code['FAT']        : 'Fat',
                nut_code['ENERC_KCAL'] : 'Energy',
                nut_code['CHOCDF'] : 'Carbohydrate',
                nut_code['FIBTG'] : 'Fiber',
                nut_code['SUGAR'] : 'Sugar'}

for name in useful_codes.values():
    df[name] = pd.Series(np.array([np.NAN]*len(df)), index=df.index) 

f = open('.\sr28\NUT_DATA.txt', 'r')
nutlines = f.readlines()
f.close()
for l in nutlines:
    tmp = l.split('~')
    if tmp[3] in useful_codes.keys():
        df.set_value(int(tmp[1]), useful_codes[tmp[3]], float(tmp[4].split('^')[1]))
```

```python
norm_mindf = mindf.copy()
cols_to_norm = [u'MG', u'P', u'K', u'NA', u'ZN', u'CU', u'SE', u'MN', u'CA', u'FE']
norm_mindf[cols_to_norm] = mindf[cols_to_norm].apply(lambda x: (x - x.mean()) / (x.max() - x.min()))
norm_mindf.MG.describe()
```

```python
center = 1. / len(norm_mindf)
norm_mindf['distance'] = pd.Series(np.array([np.NAN]*len(norm_mindf)), index=norm_mindf.index) 

def calc_dist(row):
    dist = 0
    for col in cols_to_norm:
        dist += (float(row[col])-center)**2
    return math.sqrt(dist)

for index, row in norm_mindf.iterrows():
    norm_mindf.loc[index, 'distance'] = calc_dist(row)
    
norm_mindf[['distance']][0:5]
```

```python
norm_mindf['distance'].plot(kind='hist', bins=20)
plt.text(0.8, 3000, '%s' % norm_mindf['distance'].describe())
plt.show()
```

```python
#dictionary for easy search
food_words = {}
for index, row in df.iterrows():
    fullname = row.FullName
    for w in fullname.replace(',', '').lower().split(' '):
        if w in food_words.keys():
            food_words[w].append(index)
        else:
            food_words[w] = [index]
```

```python
print food_words['salad']
```

```python
def get_foods(word):
    foods = []
    if word in food_words.keys():
        for index in food_words[word]:
            foods.append(df.loc[index, 'FullName'])
    return foods
    
print get_foods('salmon')
```
